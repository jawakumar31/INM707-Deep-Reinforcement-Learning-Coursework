```python
## Game level to vectors
dave_lvl1mat = np.zeros((8,17))
dave_lvl1mat
```

```python
dave_lvl1mat = np.zeros((8,17))
dave_lvl1mat[0,:] = -1
dave_lvl1mat[1,:] = -1
dave_lvl1mat[2,:] = -1
dave_lvl1mat[2,2::4] = 0
dave_lvl1mat[3,:] = -1
dave_lvl1mat[4,:] = -1
dave_lvl1mat[4,::4] = 0
dave_lvl1mat[5,:] = -1
dave_lvl1mat[6,:] = -1
dave_lvl1mat[6,3:7] = 0
dave_lvl1mat[6,11:16] = 0
dave_lvl1mat[7,:] = -1
dave_lvl1mat[7,11] = 0
dave_lvl1mat[7,0] = 0
# dave_lvl1mat[7,12] = 2
dave_lvl1mat
```

```python
## Destination of the Door

dave_lvl1mat[7,12] = 100
```

```python
## Defining the actions available

actions = ['right','left','jump','rightjump','leftjump','longrightjump','longleftjump']
```

```python
## Defining the states available

states = list(itertools.product(range(dave_lvl1mat.shape[0]),range(dave_lvl1mat.shape[1])))
```

```python
## Creating the R matrix

R =  np.array([[np.nan for a in actions] for s in states])
R.shape
```

```python
# R
```

```python
for i in range(len(states)):
    if (dave_lvl1mat[states[i]] == -1):
        R[i,:] = -1
    elif (dave_lvl1mat[states[i]] == 100):
        R[i,:] = 100
```

```python
## Initializing the Q matrix

Q = np.zeros(R.shape)
```

```python
## Defining the actions available

action_index = {'right':0,'left':1,'jump':2,'rightjump':3,'leftjump':4,'longrightjump':5,'longleftjump':6}
```

```python
## Defining the actions available

actions_move = {'right':(0,1),'left':(0,-1),'jump':(-2,0),'rightjump':(-2,1),'leftjump':(-2,-1),
               'longrightjump':(-3,2),'longleftjump':(-3,-2)}
```

```python
# Learning params
alpha = 0.99
gamma = 0.89
epsilon=0.9
```

```python
def get_available_actions(state):
    available_actions = []
    for i in actions_move:
        ## Checks the surrounding of the agent's state and gets you the avialable moves
        if (state[0]+actions_move[i][0]) in range(dave_lvl1mat.shape[0]) \
        and (state[1]+actions_move[i][1]) in range(dave_lvl1mat.shape[1]) \
        and dave_lvl1mat[(state[0]+actions_move[i][0],state[1]+actions_move[i][1])] != 0 :
            ## Checks if there's a wall above the agent's head to prevent it from making a jump move
            if ((dave_lvl1mat[state[0]-1,state[1]] == 0) and (i in ['jump','rightjump','leftjump',
                                                                    'longrightjump', 'longleftjump'])) :
                pass
            else:
                available_actions.append(i)
    return available_actions
```

```python
def gravity_update_state(move_state):
    ## Returns the same state if the agent is already standing on brick
    if move_state[0] == dave_lvl1mat.shape[0]-1:
        return move_state
    ## Returns the state above the brick if the agent is in a higher state from the brick
    gravity_check = [(i,move_state[1]) for i in range(move_state[0],dave_lvl1mat.shape[0])]
    for gravity in range(1,len(gravity_check)):
        if dave_lvl1mat[gravity_check[gravity]] == 0 :
            return gravity_check[gravity-1]
    return gravity_check[gravity]
```

```python
### Running the Q-Learning algorithm for 2000 episodes

dave_lvl1mat = np.zeros((8,17))
dave_lvl1mat[0,:] = -1
dave_lvl1mat[1,:] = -1
dave_lvl1mat[2,:] = -1
dave_lvl1mat[2,2::4] = 0
dave_lvl1mat[3,:] = -1
dave_lvl1mat[4,:] = -1
dave_lvl1mat[4,::4] = 0
dave_lvl1mat[5,:] = -1
dave_lvl1mat[6,:] = -1
dave_lvl1mat[6,3:7] = 0
dave_lvl1mat[6,11:16] = 0
dave_lvl1mat[7,:] = -1
dave_lvl1mat[7,11] = 0
dave_lvl1mat[7,0] = 0

dave_lvl1mat[7,12] = 100

goal_state = (7,12)

R =  np.array([[np.nan for a in actions] for s in states])

for i in range(len(states)):
    if (dave_lvl1mat[states[i]] == -1):
        R[i,:] = -1
    elif (dave_lvl1mat[states[i]] == 100):
        R[i,:] = 100
        
Q = np.zeros(R.shape)

n_episode=[]
nmoves = []
mean_reward = []

for episode in range(2000):
    n_episode.append(episode)
    state = (7,1)
    moves = 0
    for i in range(10000) :
        moves = moves+1
        available_actions = get_available_actions(state)
        available_actions

        q_values = [Q[states.index(state),action_index[a]] for a in available_actions]


        best_actions = np.array(available_actions)[np.where(q_values == np.max(q_values))[0]]


        best_actions_q_values = [Q[states.index(state),action_index[x]] for x in best_actions]


        # Epsilon-greedy
        if np.random.uniform() > epsilon:
            a = np.random.choice(available_actions)
#             print(a)
        else:
            a = np.random.choice(best_actions)
#             print(a)

        move_state = (state[0]+actions_move[a][0],state[1]+actions_move[a][1])


        action_state = gravity_update_state(move_state)


        r = R[states.index((action_state)),action_index[a]]
#         reward_sum = reward_sum + r

        q_updated = Q[states.index(state),action_index[a]] + alpha * ( r + gamma * np.max(Q[states.index(action_state),:]) - Q[states.index(state),action_index[a]])

        Q[states.index(state),action_index[a]] = q_updated
#         print(action_state)
        state = action_state
#         print(state,action_state)

        if action_state == goal_state:
            print("Ep.{} Door reached for the next level in '{}' moves. Ending episode.".format(episode,moves))
            break
    nmoves.append(moves)
    mean_reward.append(r/moves)
```

```python
# Learning params
### Experimenting the Q-Learning algorithm for diffent values of parraamters


epsilons= np.arange(0.1,1,0.1)

gammas = [0.89]*len(epsilons)
alphas = [0.99]*len(gammas)

# Learning params
alphas = [0.99]
gammas = [0.89]
epsilons = [0.9]



# epsilons= [0.9]*len(gammas)

tm_results = []

for itr in range(len(alphas)):
    print(itr)
    dave_lvl1mat = np.zeros((8,17))
    dave_lvl1mat[0,:] = -1
    dave_lvl1mat[1,:] = -1
    dave_lvl1mat[2,:] = -1
    dave_lvl1mat[2,2::4] = 0
    dave_lvl1mat[3,:] = -1
    dave_lvl1mat[4,:] = -1
    dave_lvl1mat[4,::4] = 0
    dave_lvl1mat[5,:] = -1
    dave_lvl1mat[6,:] = -1
    dave_lvl1mat[6,3:7] = 0
    dave_lvl1mat[6,11:16] = 0
    dave_lvl1mat[7,:] = -1
    dave_lvl1mat[7,11] = 0
    dave_lvl1mat[7,0] = 0

    dave_lvl1mat[7,12] = 100

    goal_state = (7,12)

    R =  np.array([[np.nan for a in actions] for s in states])

    for i in range(len(states)):
        if (dave_lvl1mat[states[i]] == -1):
            R[i,:] = -1
        elif (dave_lvl1mat[states[i]] == 100):
            R[i,:] = 100

    Q = np.zeros(R.shape)

    n_episode=[]
    nmoves = []
    mean_reward = []

    for episode in range(1000):
        n_episode.append(episode)
        state = (7,1)
        moves = 0
        for i in range(10000) :
            moves = moves+1
            available_actions = get_available_actions(state)
            available_actions

            q_values = [Q[states.index(state),action_index[a]] for a in available_actions]


            best_actions = np.array(available_actions)[np.where(q_values == np.max(q_values))[0]]


            best_actions_q_values = [Q[states.index(state),action_index[x]] for x in best_actions]


            # Epsilon-greedy
            if np.random.uniform() > epsilons[itr]:
                a = np.random.choice(available_actions)
    #             print(a)
            else:
                a = np.random.choice(best_actions)
    #             print(a)

            move_state = (state[0]+actions_move[a][0],state[1]+actions_move[a][1])


            action_state = gravity_update_state(move_state)


            r = R[states.index((action_state)),action_index[a]]
    #         reward_sum = reward_sum + r

            q_updated = Q[states.index(state),action_index[a]] + alphas[itr] * ( r + gammas[itr] * np.max(Q[states.index(action_state),:]) - Q[states.index(state),action_index[a]])

            Q[states.index(state),action_index[a]] = q_updated
    #         print(action_state)
            state = action_state
    #         print(state,action_state)

            if action_state == goal_state:
                print("Ep.{} Door reached for the next level in '{}' moves. Ending episode.".format(episode,moves))
                break
        nmoves.append(moves)
        mean_reward.append(r/moves)
        
        tm_results.append([itr,episode,moves,r/moves])
```

```python
temp_df = pd.DataFrame(tm_results,columns=['iter','episode','nmoves','rewards'])
```
